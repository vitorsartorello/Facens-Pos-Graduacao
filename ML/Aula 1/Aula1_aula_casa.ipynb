Dentre todas as categorias de fatores relacionados à saúde, foram selecionados apenas os fatores críticos mais representativos. Observou-se que, nos últimos 15 anos, houve um grande desenvolvimento no setor de saúde, resultando na redução das taxas de mortalidade humana, especialmente nos países em desenvolvimento, em comparação com os últimos 30 anos. # **Explicação das Colunas**


* Country - País                        
* Year - Idade                               
* Status - Status de país desenvolvido ou em desenvolvimento                            
* Life expectancy - Expectativa de vida por idade               
* Adult Mortality - Taxa de mortalidade adulta para ambos os sexos (probabilidade de morte entre 15 e 60 anos por 1000 habitantes)         
* infant deaths - Número de mortes infantis por 1000 habitantes
* Alcohol - Álcool, consumo per capita (15+) registado (em litros de álcool puro)
* percentage expenditure - Percentual de Gastos, despesas com saúde como porcentagem do Produto Interno Bruto per capita (%)
* Hepatitis B - Cobertura vacinal contra hepatite B (HepB) em crianças de 1 ano de idade (%)
* Measles - Sarampo, número de casos notificados por 1000 habitantes.
* BMI - IMC, Índice de Massa Corporal médio de toda a população
* under-five deaths - Número de mortes de crianças menores de cinco anos por 1000 habitantes
* Polio - Cobertura vacinal contra a poliomielite (Pol3) em crianças de 1 ano de idade (%)   
* Total expenditure - Despesas do governo geral com saúde como porcentagem das despesas totais do governo (%)
* Diphtheria - Cobertura da vacinação contra difteria, tétano e coqueluche (DTP3) em crianças de 1 ano de idade (%)
* HIV/AIDS - Mortes por 1.000 nascidos vivos por HIV/AIDS (0-4 anos)  
* GDP - PIB, Produto Interno Bruto per capita (em USD)  
* Population - População do País
* thinness  1-19 years - Magreza 10 - 19 anos, prevalência de magreza entre crianças e adolescentes de 10 a 19 anos (%)  
* thinness 5-9 years - Prevalência de magreza em crianças de 5 a 9 anos (%)     
* Income composition of resources - Índice de Desenvolvimento Humano em termos de composição da renda dos recursos (índice variando de 0 a 1)  
* Schooling - Escolarização, número de anos de escolaridade (anos)

# **1. Análise Exploratória Geral**

Responda às seguintes perguntas:


1.   Quantas linhas e colunas temos no nosso dataset?
2.   Temos dados nulos? Se sim, qual a porcentagem deles? É uma porcentagem significativa?
3. Sobre os dados nulos com maior porcentagem, qual o tipo deles? Tente pensar em hipóteses do porquê eles estão nulos.
4. Existem duplicatas?
5. Todas as colunas estão com o seu tipo correto?
6. Como verificar as principais estatísticas das colunas numéricas? Escolha uma coluna com valores nulos para verificar como a média, moda, mediana.




```python
import pandas as pd
import numpy as np

# ===== Carregamento =====
df = pd.read_csv("expectancy.csv")

print("=== 1) Linhas e colunas ===")
print(f"shape: {df.shape}  ->  linhas={df.shape[0]}, colunas={df.shape[1]}")
print()

print("=== 2) Nulos por coluna (contagem e %) ===")
null_count = df.isna().sum()
null_pct = df.isna().mean() * 100
display(pd.DataFrame({"nulos": null_count, "%": null_pct}).sort_values("%", ascending=False))

overall_null_pct = (df.isna().sum().sum() / (df.shape[0] * df.shape[1])) * 100
print(f"\n% nulos no dataset inteiro: {overall_null_pct:.2f}%")
print()

print("=== 3) Colunas com maior % de nulos e seus tipos ===")
top_n = 5  # ajuste se quiser ver mais/menos
top_null_cols = null_pct.sort_values(ascending=False).head(top_n).index.tolist()
print("Top colunas com nulos:", top_null_cols)
print("\nTipos das top colunas com nulos:")
print(df[top_null_cols].dtypes)
# (Hipóteses sobre 'por que' ficam na análise textual do seu relatório.)
print()

print("=== 4) Duplicatas ===")
dup_count = df.duplicated().sum()
print(f"Linhas duplicadas: {dup_count}")
print()

print("=== 5) Tipos das colunas (checagem) ===")
print(df.dtypes)
# (Opcional) detectar objetos que parecem numéricos:
obj_cols = df.select_dtypes(include="object").columns
suspeitas_numericas = []
for c in obj_cols:
    coerced = pd.to_numeric(df[c], errors="coerce")
    # Se ao converter >80% vira número, é forte candidato a ser numérico:
    if coerced.notna().mean() > 0.8:
        suspeitas_numericas.append(c)
if suspeitas_numericas:
    print("\nPossíveis colunas 'object' que parecem numéricas (verificar conversão):")
    for c in suspeitas_numericas:
        print(" -", c)
print()

print("=== 6) Estatísticas de uma coluna com nulos ===")
# escolhe automaticamente a coluna com maior % de nulos (>0)
cols_com_nulos = null_pct[null_pct > 0].sort_values(ascending=False)
if not cols_com_nulos.empty:
    col = cols_com_nulos.index[0]  # pegue uma com nulos (a de maior %)
    serie = pd.to_numeric(df[col], errors="coerce") if df[col].dtype == "object" else df[col]
    print(f"Coluna escolhida: {col}")
    print(f"count (não-nulos): {serie.notna().sum()}")
    print(f"mean : {serie.mean():.4f}")
    print(f"median: {serie.median():.4f}")
    moda = serie.mode(dropna=True)
    if len(moda) > 0:
        print(f"mode : {', '.join([str(m) for m in moda.values[:3]])}  (mostrando até 3)")
    else:
        print("mode : (sem moda definida)")
else:
    print("Não há colunas com valores nulos para analisar.")
```

```python
# 13. Países mais ricos vivem mais?
plt.figure(figsize=(8,5))
# Filtrar linhas com GDP e Life expectancy válidos
df_filtrado = df[(df['GDP'] > 0) & (df['Life expectancy'] > 0)].dropna(subset=['GDP', 'Life expectancy'])
plt.xscale('log')
sns.scatterplot(x='GDP', y='Life expectancy', data=df_filtrado)
plt.title('PIB (escala log) x Expectativa de Vida')
plt.show()
 
```

```python
# 15. Países desenvolvidos consomem mais álcool?

plt.figure(figsize=(8,5))
sns.boxplot(x='Status', y='Alcohol', data=df)
plt.title('Consumo de Álcool por Status de Desenvolvimento')
plt.xlabel('Status do País')
plt.ylabel('Consumo de Álcool (litros per capita)')
plt.show()

# Cálculo das médias
medias = df.groupby('Status')['Alcohol'].mean()

# Justificativa formatada
media_dev = medias['Developed']
media_devp = medias['Developing']

print(f"Média de consumo de álcool (litros per capita):")
print(f" - Países Desenvolvidos: {media_dev:.2f}")
print(f" - Países em Desenvolvimento: {media_devp:.2f}\n")

print(f"Países desenvolvidos consomem, em média, mais álcool que os em desenvolvimento.")
print(f"Isto pode estar relacionado ao maior poder aquisitivo e acesso a bebidas alcoólicas,")
print(f"além de fatores culturais e econômicos que influenciam o consumo.")
```

```python
#17. Como a expectativa de vida evoluiu ao longo do tempo?
# Evolução da expectativa de vida ao longo do tempo (média global)
df['Year'] = df['Year'].astype(str).str.replace('[oO]', '0', regex=True)
 
df_mean = df.groupby('Year', as_index=False)['Life expectancy'].mean()
 
plt.figure(figsize=(8, 5))
sns.lineplot(x='Year', y='Life expectancy', data=df_mean, marker='o', color='teal')
 
plt.title('Evolução da Expectativa de Vida ao Longo dos Anos')
plt.xlabel('Ano')
plt.ylabel('Expectativa de Vida (anos)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

```

```python
#18. Países com maior investimento em saúde têm maior expectativa de vida?

plt.figure(figsize=(8,5))
sns.scatterplot(x='percentage expenditure', y='Life expectancy', data=df)
plt.xscale('log')
plt.title('% de Investimento em Saúde (log) x Expectativa de Vida')
plt.xlabel('% de Investimento em Saúde (escala log)')
plt.ylabel('Expectativa de Vida (anos)')
plt.show()
 
```
